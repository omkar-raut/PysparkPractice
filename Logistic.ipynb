```
data=df2.select(['Age',
 'Total_Purchase',
 'Account_Manager',
 'Years',
 'Num_Sites',
 'Churn'])
```

```
from pyspark.ml.feature import (VectorAssembler,VectorIndexer,
                                OneHotEncoder,StringIndexer)
```

```
data.show()
```

```
LocationIndex=StringIndexer(inputCol='Location',outputCol='LocationIndex')
LocationEncoder=OneHotEncoder(inputCol='LocationIndex',outputCol='LocationEncoded')
```

```
assembler=VectorAssembler(inputCols=['Age',
 'Total_Purchase',
 'Account_Manager',
 'Years',
 'Num_Sites'],outputCol='features')
final_data=assembler.transform(data).select('Churn','features')
```

```
final_data.show()
```

```
from pyspark.ml.classification import LogisticRegression
```

```

train,test=final_data.randomSplit([0.7,0.3])
test.show()
```

```
model_churn = LogisticRegression(featuresCol='features',labelCol='Churn')
```

```
resuls=model_churn.fit(dataset=train).transform(dataset=test)
resuls.show()
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
my_eval = BinaryClassificationEvaluator(rawPredictionCol='prediction',
                                       labelCol='Churn')
```

```
resuls.show()
```

```
fnl=my_eval.evaluate(resuls)
```
